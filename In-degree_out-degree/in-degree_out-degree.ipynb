### Computing In-degree and Out-degree for directed graph

```python
f = open("metabolic.edgelist.txt")
```

```python
list_ =  f.readlines()
```

```python
u = []
v = []

for i in list_:
    getArray = i.split("\t")
    
    getU = getArray[0]
    getV = getArray[1][:-1]
    
    u.append(int(getU))
    v.append(int(getV))
```

```python
indegree = {}

for item in u:
    if (item in indegree):
        indegree[item] += 1
    else:
        indegree[item] = 1

indegree = dict(sorted(indegree.items()))        
        
print(indegree)
```

```python
outdegree = {}

for item in v:
    if (item in outdegree):
        outdegree[item] += 1
    else:
        outdegree[item] = 1

outdegree = dict(sorted(outdegree.items()))                
        
print(outdegree)
```

```python
import matplotlib.pyplot as plt
```

```python
#indegree
Xval = list(indegree.keys())
Yval = list(indegree.values())

plt.xlabel('Nodes')
plt.ylabel('Degree')
plt.plot(Xval,Yval)
plt.title('Indegree')
```

```python
#outdegree

Xval = list(outdegree.keys())
Yval = list(outdegree.values())

plt.xlabel('Nodes')
plt.ylabel('Degree')
plt.plot(Xval,Yval)
plt.title('Outdegree')
```
